```
from pyspark import SparkConf,SparkContext
conf = SparkConf().setAppName('Mini Project')
sc = SparkContext.getOrCreate(conf = conf)
```

```
rdd = sc.textFile('/FileStore/tables/students___Sheet1.csv')
headers = rdd.first()
rdd = rdd.filter(lambda x : x!=headers)
rdd = rdd.map(lambda x: x.split(','))

```

```
rdd.count()
#no of students is 9 in the file #Show no of students in the file
```

```
#Total Marks achived by Female and Male students 
rdd2 = rdd
rdd2 = rdd.map(lambda x : (x[1], int(x[5])))
rdd2 = rdd2.reduceByKey(lambda x,y : x+y)
rdd2.collect()
```

```
#Total no students passed and failed 50+ marks means pass 
rdd3 = rdd
passed = rdd3.filter(lambda x : int(x[5]) > 50 ).count() #no of pass stundents in the data set 
Failed = rdd3.filter(lambda x : int(x[5]) < 50 ).count() #no of fail stundents in the data set 
print(passed,Failed)

#menthod 2

passed2 = rdd3.filter(lambda x : int(x[5]) > 50 ).count() #no of pass stundents in the data set
failed2 =rdd.count()-passed2
print(passed2,failed2)
```

```
#Show Total no of students enrolled per course 
rdd4 = rdd 
rdd4 = rdd4.map(lambda x: (x[3],1))
rdd4.reduceByKey(lambda x,y : x+ y).collect()
```

```
#No of Marks by course 
rdd5 = rdd
rdd5 = rdd5.map(lambda x:(x[3],int(x[5])))
rdd5.reduceByKey(lambda x,y : x+y).collect()
```

```
#Average marks achived in course by the  students 
rdd6 = rdd
rdd6 = rdd6.map(lambda x: (x[3], (int(x[5]),1)))
rdd6 = rdd6.reduceByKey(lambda x,y:  (x[0]+y[0],x[1]+y[1]))
rdd6.map(lambda x : (x[0],x[1][0]/x[1][1])).collect()
```

```
#Useing Map Value
rdd6.mapValues(lambda x : (x[0]/x[1])).collect()
```

```
#min and max marks achived per course 
rdd7 = rdd 
rdd7 = rdd7.map(lambda x : (x[3],int(x[5]) ))
maxx = rdd7.reduceByKey(lambda x,y : x if x> y else y)
minn = rdd7.reduceByKey(lambda x,y : x if x< y else y)
maxx.collect()
minn.collect()
```
